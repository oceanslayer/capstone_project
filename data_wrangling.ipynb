```python
print(df['LaunchSite'].value_counts())
print('-------------------------')
print(df['Orbit'].value_counts())
print('-------------------------')
landing_outcomes = df['Outcome'].value_counts()
print(landing_outcomes)
```

```python
bad_outcomes=set(landing_outcomes.keys()[[1,3,5,6,7]])

landing_class = []
for i in range(len(df['Outcome'])):
    if df['Outcome'].loc[i] in bad_outcomes:
        landing_class.append(0)
    else:
        landing_class.append(1)

df['Class']=landing_class
success_rate_str = str((round(df["Class"].mean(), 3))*100) + str('%')
print('The average success rate of outcome is', success_rate_str)
```
