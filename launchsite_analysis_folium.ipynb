```python
import folium
import pandas as pd
from folium.plugins import MarkerCluster
from folium.features import DivIcon
```

```python
df=pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv')
df.head()
```

```python
df = df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python
# Start location is NASA Johnson Space Center at Houston, Texas
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=12)
```

```python
circle = folium.Circle(nasa_coordinate, radius=1000, color='yellow', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
marker = folium.map.Marker(nasa_coordinate, icon=DivIcon(icon_size=(20,20), icon_anchor=(0,0), html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC'))
site_map.add_child(circle)
site_map.add_child(marker)
```

```python
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

for i in range(len(launch_sites_df)):
    circle = folium.Circle([launch_sites_df.loc[i,'Lat'],launch_sites_df.loc[i,'Long']], radius=1000, color='yellow', fill=True).add_child(folium.Popup(str(launch_sites_df.loc[i,'Launch Site'])))
    marker = folium.map.Marker([launch_sites_df.loc[i,'Lat'],launch_sites_df.loc[i,'Long']], icon=DivIcon(icon_size=(20,20), icon_anchor=(0,0), html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % str(launch_sites_df.loc[i,'Launch Site'])))
    site_map.add_child(circle)
    site_map.add_child(marker)
site_map
```

```python
marker_cluster = MarkerCluster()
```

```python
# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
df['marker_color'] = df['class'].apply(assign_marker_color)
df.tail(10)
```
